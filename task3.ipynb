```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

# Path to the dataset folders
data_path = '/kaggle/input/eeg-signals-impulse/train_data-20250117T173905Z-001/train_data'
folders = ['Normal', 'Complex_Partial_Seizures', 'Electrographic_Seizures', 'Video_detected_Seizures_with_no_visual_change_over_EEG']
```

```python
def load_eeg_data(folder):
    # Create the full folder path
    folder_path = os.path.join(data_path, folder)
    print(f"Checking folder path: {folder_path}")

    # Check if the folder exists
    if not os.path.exists(folder_path):
        print(f"Folder {folder_path} does not exist.")
        return None

    # List all files in the folder
    files = os.listdir(folder_path)
    print(f"Files in {folder}: {files}")

    # Check if the folder is empty
    if not files:
        print(f"Folder {folder_path} is empty.")
        return None

    # Get the first file and check if it's a .npy file
    file_path = os.path.join(folder_path, files[0])
    if not file_path.endswith('.npy'):
        print(f"Skipping {file_path} as it is not a .npy file.")
        return None

    # Attempt to load the .npy file
    print(f"Loading data from: {file_path}")
    try:
        data = np.load(file_path)
        print(f"Loaded data shape: {data.shape}")
        return data
    except Exception as e:
        print(f"Error loading data from {file_path}: {e}")
        return None

# Iterate through the folders and load one file from each
for folder in folders:
    print(f"Processing folder: {folder}")
    data = load_eeg_data(folder)
    if data is not None:
        print(data)  # Print the data (or process it further)
    else:
        print(f"No data loaded from folder: {folder}")
```

```python
import os
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler
from scipy.fft import fft
```

```python
#mapped class labels based on folder names.

train_data_path = "/kaggle/input/eeg-signals-impulse/train_data-20250117T173905Z-001/train_data"
val_data_path = "/kaggle/input/val-data//kaggle/input/val-data/validation_data/"

class_labels = {
    "Complex_Partial_Seizures": 0,
    "Electrographic_Seizures": 1,
    "Normal": 2,
    "Video_detected_Seizures_with_no_visual_change_over_EEG": 3
}
```

```python
x_train = []
y_train = []

for class_name, label in class_labels.items():
    class_folder = os.path.join(train_data_path, class_name)
    for file_name in os.listdir(class_folder):
        file_path = os.path.join(class_folder, file_name)
        if file_name.endswith('.npy'):
            data = np.load(file_path)
            x_train.append(data)
            y_train.append(label)
```

```python
x_val = []
y_val = []

for class_name, label in class_labels.items():
    class_folder = os.path.join(val_data_path, class_name)
    for file_name in os.listdir(class_folder):
        file_path = os.path.join(class_folder, file_name)
        if file_name.endswith('.npy'):
            data = np.load(file_path)
            x_val.append(data)
            y_val.append(label)
```

```python
x_train = np.array(x_train)
y_train = np.array(y_train)

print(f"x_train shape: {x_train.shape}")
print(f"y_train shape: {y_train.shape}")
```

```python
#
import numpy as np

def generate_features(signals, n):
    """
    Generate comprehensive features for a list of EEG signals.

    Parameters:
        signals (list of np.ndarray): List of 1D arrays representing EEG signals.
        n (int): Number of Fourier coefficients to extract.

    Returns:
        np.ndarray: 2D array where each row corresponds to the feature set of a signal.
    """
    def fourier_features(signal, n):
        """Compute the first n Fourier coefficients (magnitude)."""
        fft = np.fft.fft(signal)
        return np.abs(fft[:n])

    def zero_crossing_rate(signal):
        """Calculate the zero-crossing rate of a signal."""
        return ((signal[:-1] * signal[1:]) < 0).sum() / len(signal)

    def statistical_features(signal):
        """Compute statistical features like mean, variance, skewness, and kurtosis."""
        mean = np.mean(signal)
        std = np.std(signal)
        var = np.var(signal)
        skewness = np.mean((signal - mean) ** 3) / (std ** 3 + 1e-8)  # Avoid division by zero
        kurtosis = np.mean((signal - mean) ** 4) / (std ** 4 + 1e-8)
        return np.array([mean, std, var, skewness, kurtosis])

    def energy_and_rms(signal):
        """Compute signal energy and root mean square."""
        energy = np.sum(signal ** 2)
        rms = np.sqrt(np.mean(signal ** 2))
        return np.array([energy, rms])

    features = []
    for signal in signals:
        if len(signal) == 0:
            raise ValueError("Empty signal encountered.")

        # Frequency-domain features
        ff = fourier_features(signal, n)

        # Time-domain features
        zcr = zero_crossing_rate(signal)
        stats = statistical_features(signal)
        energy_rms = energy_and_rms(signal)

        # Combine all features
        combined_features = np.concatenate((ff, [zcr], stats, energy_rms))
        features.append(combined_features)

    # Convert list of features to a 2D NumPy array
    feature_array = np.array(features)

    # Normalize features (optional but recommended for machine learning models)
    normalized_features = (feature_array - np.mean(feature_array, axis=0)) / (np.std(feature_array, axis=0) + 1e-8)

    return normalized_features


features = generate_features(x_train, n)
features_val = generate_features(x_val, n)


print(f"Training features shape: {features.shape}")
print(f"Validation features shape: {features_val.shape}")

if features.shape[1] != features_val.shape[1]:
    min_features = min(features.shape[1], features_val.shape[1])
    features = features[:, :min_features]
    features_val = features_val[:, :min_features]

scaler = StandardScaler()
features = scaler.fit_transform(features)
features_val = scaler.transform(features_val)


svm_model = SVC(kernel='rbf')
svm_model.fit(features, y_train)

y_pred = svm_model.predict(features_val)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, balanced_accuracy_score
from sklearn.preprocessing import label_binarize

# Predict on validation set
y_pred = svm_model.predict(features_val)

print("Accuracy:", accuracy_score(y_val, y_pred))
balanced_acc = balanced_accuracy_score(y_val, y_pred)
print(f"\nBalanced Accuracy: {balanced_acc:.4f}")

# ROC AUC (for multi-class classification, binarize labels)
y_val_bin = label_binarize(y_val, classes=np.unique(y_val))
y_pred_bin = label_binarize(y_pred, classes=np.unique(y_val))

roc_auc = roc_auc_score(y_val_bin, y_pred_bin, average='macro', multi_class='ovr')  # one-vs-rest
print(f"\nROC AUC Score: {roc_auc:.4f}")
```
